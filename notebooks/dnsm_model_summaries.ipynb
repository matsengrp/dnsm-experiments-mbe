```python
import pandas as pd

from dnsmex import dnsm_zoo

from dnsmex.dxsm_model_summaries import build_summary_df
```

```python
summary_df = build_summary_df(dnsm_zoo)
summary_df.to_csv("../output/dnsm_model_parameter_count.csv", index=False)

summary_df
```

```python
# drop model from summary_df
df = summary_df.drop(columns=["model"])
df = df.drop([1, 2, 7, 8])
rename_dict = {
    "abbrv_param_count": "Label",
    "head_count": "Heads",
    "d_model_per_head": "Model dim per head",
    "layer_count": "Layers",
    "parameters": "Parameters"
}
# rename columns
df = df.rename(columns=rename_dict)
# replace the Parameters column with a string version with commas
df["Parameters"] = df["Parameters"].apply(lambda x: f"{x:,}")
df.to_latex("../output/model_parameter_count.tex", index=False, escape=False)
```
